```python
def rock_paper_scissors():
    print('¡Hola amigo!\n\n>>>Welcome to the game of ROCK-PAPER-SCISSORS<<<\n\nThis game is designed to be played against the Computer.')
    print('The one with the highest score will win.')
    print('___________________________________________________________')
    player_name = input('Please enter your name: ')
    def rps_gameplay():
        try:
            turns = int(input(f'Hello {player_name}, How many rounds would you like to play? '))
            from random import randint
            i = 1
            player_score = 0
            computer_score = 0
            while i <= turns:
                p = (input(f'Round {i} of {turns}. Select 1 for Rock,2 for Paper or 3 for Scissors? >>>'))
                if p not in ['1','2','3']:
                    print("You've selected an invalid choice. Please try again")
                else:
                    player_input = ('ROCK' if p == '1' else 'PAPER' if p == '2' else 'SCISSORS')
                    x = randint(1,3)
                    computer_input = ('ROCK' if x == 1 else 'PAPER' if x == 2 else 'SCISSORS')
                    print('___________________________________________________________')
                    print(f"Your Choice --> {player_input}\nComputer's Choice --> {computer_input}")
                    if player_input == computer_input:
                        print(">>>It's a Tie<<<")
                        print(f"SCORE --> {player_name}-{player_score}|Computer-{computer_score}")
                        print('___________________________________________________________')
                    else:
                        if player_input == 'ROCK' and computer_input == 'SCISSORS':
                            print(">>>You've won this round!<<<")
                            player_score += 1
                            print(f"SCORE --> {player_name}-{player_score}|Computer-{computer_score}")
                            print('___________________________________________________________')
                        elif player_input == 'PAPER' and computer_input == 'ROCK':
                            print(">>>You've won this round!<<<")
                            player_score += 1
                            print(f"SCORE --> {player_name}-{player_score}|Computer-{computer_score}")
                            print('___________________________________________________________')
                        elif player_input == 'SCISSORS' and computer_input == 'PAPER':
                            print(">>>You've won this round!<<<")
                            player_score += 1
                            print(f"SCORE --> {player_name}-{player_score}|Computer-{computer_score}")
                            print('___________________________________________________________')
                        else:
                            print(">>>Computer has won this round<<<")
                            computer_score += 1
                            print(f"SCORE --> {player_name}-{player_score}|Computer-{computer_score}")
                            print('___________________________________________________________')
                    i += 1
            print('=======================================================================')
            if player_score == computer_score:
                print(f"The game was a tie with a score of {player_score}-{computer_score}")
            elif player_score > computer_score:
                print(f"Congratulations! You've won this game with a score of {player_score}-{computer_score}")
            else:
                print(f"Better luck next time! You've lost to the Computer with a score of {player_score}-{computer_score}")
            print('=======================================================================')
        
            play = input('Do you want to play again? Type "Y" for Yes & "N" for No.>>> ').upper()
            if play == "Y":
                rps_gameplay()
            elif play == "N":
                print('Thank you for playing! Have a good day!')
            else:
                print('Invalid Choice! Adios!')
        except ValueError:
            print("\n###The turns has to be a number. Let's start again!###\n")
            rps_gameplay()
    rps_gameplay()
rock_paper_scissors()
```
